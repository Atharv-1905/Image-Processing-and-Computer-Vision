```python
img1 = cv2.imread('/content/download.jpeg')
plt.imshow(img1)
```

```python
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img3 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img3)
```

```python
img4 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img4, cmap='gray')
```

```python
height, width, _ = img2.shape
img_resized = cv2.resize(img2, (width // 2, height // 2), interpolation=cv2.INTER_AREA)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2)
plt.title('Original Image (img2)')
plt.subplot(1, 2, 2)
plt.imshow(img_resized)
plt.title('Resized (INTER_AREA)')
plt.show()
```

```python
img_blurred = cv2.GaussianBlur(img2, (25, 25), 0)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2)
plt.title('Original Image (img2)')
plt.subplot(1, 2, 2)
plt.imshow(img_blurred)
plt.title('Blurred Image')
plt.show()
```

```python
height, width, _ = img2.shape
center = (width // 2, height // 2)

# Get the rotation matrix without adjusting for new dimensions
M_simple = cv2.getRotationMatrix2D(center, 45, 1.0)

# Perform the rotation within original dimensions
rotated_img_simple = cv2.warpAffine(img2, M_simple, (width, height))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2)
plt.title('Original Image (img2)')
plt.subplot(1, 2, 2)
plt.imshow(rotated_img_simple)
plt.title('Rotated Image (45 degrees, simpler)')
plt.show()
```

```python
height, width, _ = img2.shape
center = (width // 2, height // 2)

# Get the rotation matrix
M = cv2.getRotationMatrix2D(center, 45, 1.0)

# Calculate new image dimensions to fit the entire rotated image
abs_cos = abs(M[0, 0])
abs_sin = abs(M[0, 1])
new_width = int(height * abs_sin + width * abs_cos)
new_height = int(height * abs_cos + width * abs_sin)

# Adjust the translation component of the rotation matrix
M[0, 2] += (new_width / 2) - center[0]
M[1, 2] += (new_height / 2) - center[1]

# Perform the rotation
rotated_img = cv2.warpAffine(img2, M, (new_width, new_height))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2)
plt.title('Original Image (img2)')
plt.subplot(1, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated Image (45 degrees)')
plt.show()
```
